```python
from pyspark.sql import SparkSession
spark = SparkSession\
        .builder\
        .enableHiveSupport() \
        .appName("IGNGames")\
        .getOrCreate()
```

```python
games= spark.read.csv("D:\RomiCode\DataScience\Datasets\GyanMatrix-IGN\ign.csv",inferSchema="True",header="True")
```

```python
games.createOrReplaceTempView("games")
```

```python

spark.sql("select * from games").show()
```

```python
spark.sql("select distinct score_phrase from games").show()
```

```python
spark.sql("select distinct score from games").show()
```

```python
spark.sql("select distinct score,score_phrase from games  order by 1 desc").collect()
```

```python
spark.sql("select distinct score,score_phrase,floor(score) from games  order by 1 desc").collect()
```

```python
spark.sql("select distinct platform from games  order by 1 desc").collect()
```

```python
spark.sql("select distinct genre from games  order by 1 desc").collect()
```

```python
spark.sql("select * from games where editors_choice = 'Y' order by 1 desc").show()
```

```python
spark.sql("select count(*),  genre, editors_choice from games where editors_choice = 'Y' group by  genre, editors_choice order by 1 desc").collect()
```

```python
spark.sql("select count(*), release_month, genre, editors_choice from games where editors_choice = 'Y' group by  release_month, genre, editors_choice order by 1 desc").collect()
```

```python
spark.sql("select count(*), concat ('Q', ceil(release_month/3)) as Quarter, genre, editors_choice from games where editors_choice = 'Y' group by  ceil(release_month/3), genre, editors_choice order by 1 desc").collect()
```

```python
spark.sql("select count(*), concat ('Q', ceil(release_month/3)) as Quarter, editors_choice from games where editors_choice = 'Y' group by  ceil(release_month/3),  editors_choice order by 1 desc").collect()
```

```python
spark.sql("select count(*), concat ('Q', ceil(release_month/3)) as Quarter, editors_choice from games group by  ceil(release_month/3),  editors_choice order by 2 asc, 3 desc").collect()
```

```python
spark.sql("select genre from games where editors_choice = 'Y' minus select genre from games where editors_choice = 'N'").collect()
```

```python
spark.sql("select genre from games where editors_choice = 'N' minus select genre from games where editors_choice = 'Y'").collect()
```

```python
spark.sql("select * from games where editors_choice = 'Y' and genre = 'Educational'").collect()
```

```python
spark.sql("select * from games where editors_choice = 'Y' and genre like '%Action%'").collect()
```
